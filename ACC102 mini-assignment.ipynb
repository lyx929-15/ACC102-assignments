```python
import pandas as pd
import matplotlib.pyplot as plt
import wrds
```

```python
username = "lilya"
db = wrds.Connection(wrds_username=username)
```

```python
ticker = "TSLA"      
start_date = "2020-01-01"
end_date="2023-12-31"

selected_columns = """
b.htsymbol,
b.permno,
a.date,
a.prc,
a.ret
"""
```

```python
sql_query = f"""
SELECT {selected_columns}
FROM crsp.msf AS a
LEFT JOIN crsp.msfhdr AS b
    ON a.permno = b.permno
WHERE b.htsymbol = '{ticker}'
AND a.date >= '{start_date}'
AND a.date <= '{end_date}'
"""
df = db.raw_sql(
    query, 
    params={
        "ticker": ticker, 
        "start_date": start_date, 
        "end_date": end_date    
    },
    date_cols=["date"]
)
print(sql_query)
df
```

```python
df_sorted = df.sort_values('date')

plt.figure(figsize=(12, 6))

plt.plot(df_sorted['date'], df_sorted['prc'], 
         linewidth=2, color='blue', label='TSLA Price')

plt.title('Tesla Stock Price Over Time', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Price (cents) /100', fontsize=12)  
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```
